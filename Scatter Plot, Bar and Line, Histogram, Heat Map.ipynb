# Scatter Plot

In a scatter plot, the values of two variables are plotted as points on a two dimensional grid.

Additionally, a third variable can be used to determine the size or color of the points.

The Iris Flower Dataset has been used for the following visualization and the same has been included in sns lib.

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
type(df)
```

```python
list(pd.unique(df.species)) #unique species in the dataset
```

Let's firt try to visualize the relationship between sepal length and sepal width.

Our first instinct might be to create a line chart using plt.plot.

```python
sns.set_style("whitegrid")
plt.plot(df.sepal_length,df.sepal_width);
```

Lets try using a scatter plot instead, and see how good the visualization is

```python
sns.scatterplot(df.sepal_length,df.sepal_width);
```

```python
sns.scatterplot(df.sepal_length,
                df.sepal_width,
                hue=df.species,
                s=100);
```

Adding hue makes the plot more informative.
We can immediately tell that flowers of the setosa have a lower sepal length and a larger sepal width in comparison

Customizing Seaborn Figures

Fact: 
--Seaborn uses Matplotlib's function internally.
--As a result, we can use functions like plt.figure,plt.title to modify the figure

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')

sns.scatterplot(df.sepal_length,
                df.sepal_width,
                hue=df.species,
                s=100);
```

The other way of doing the same as above

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')

sns.scatterplot('sepal_length',
                'sepal_width',
                hue=df.species,
                s=100,
                data = df);
```

# Histogram

Histograms are used to represent a single series or list of data.

It creates bins and represents the frequency of the occurance as bar graphs.

In other terms, histogram is a one variable chart with continious bars in nature

```python
plt.title("Distribution of Sepal Width")
plt.hist(df.sepal_width);
```

Most observations lie between sepal width of 3.0 to 3.5

Controlling Size and Number of Bins

```python
plt.hist(df.sepal_width,bins=5);
```

To provide the interval, we use np.arange() function from numpy and pass it through plt.hist() function

```python
import numpy as np

np.arange(2,5,0.25)
```

```python
plt.hist(df.sepal_width,bins=np.arange(2,5,0.25));
```

```python
#we can also have bins of unequal sizes
plt.hist(df.sepal_width,bins=[1,3,4.5]);
```

Multiple Histograms

Similar to line charts, we can draw multiple histograms in a single chart.

We can reduce the opacity of each histogram, so that the bars of one histogram dont hide the bars for others.

Drawing histograms for each species of flowers

```python
setosa_df=df[df.species=='setosa']
versicolor_df=df[df.species=='versicolor']
virginica_df=df[df.species=='virginica']
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
```

Adding legend for readability

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));

plt.legend(['Setosa','Versicolor'])
#Be careful when declaring the legend. 
#The first order in which series are declared in .hist() should be followed,
#whn declaring legend names 
```

Stacking multiple histograms on top of one another

```python
#When stack is True
plt.title("Distribution of Sepal Width")

plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],
         bins=np.arange(2,5,0.25),
        stacked=True);

plt.legend(['Setosa','Versicolor','Virginica'])
```

```python
#When stack is false
plt.title("Distribution of Sepal Width")

plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],
         bins=np.arange(2,5,0.25),
        stacked=False);

plt.legend(['Setosa','Versicolor','Virginica']);
```

# Bar Charts

```python
year = [2015,2016,2017,2018,2019,2020]
yield_apples = [340,230,600,590,650,720]
yield_oranges = [350,380,400,380,420,410]
```

```python
sns.set_style("darkgrid")
plt.plot(year,yield_apples);
```

```python
plt.bar(year,yield_apples)
plt.title("Bar chart showing the yield of apples")
plt.legend(['Apples']);
```

Drawing both line and bar chart in the same graph

```python
plt.bar(year,yield_apples)
plt.plot(year,yield_apples, 'o--r')
plt.title("Bar chart showing the yield of apples")
plt.legend(['Apples','Apples']);
```

There is a small sublity involved.

We know that the legend should be declared in the same order as the chart.

But here, irrespective of order, line chart is given the first prioirty and the bar later.

As both are apples here. It is not a concern. But make sure you habe understood the order of how pyplot works

Lets try bar with oranges and line with apple

```python
plt.bar(year,yield_oranges)
plt.plot(year,yield_apples, 'o--r')
plt.title("Bar and line in the same chart showing the yield of oranges and apples respectively")
plt.legend(['Oranges','Apples']);
```

CAUTION: Note the line should be orange and bar the apples.

Redeclare the legend

```python
plt.bar(year,yield_oranges)
plt.plot(year,yield_apples, 'o--r')
plt.title("Bar and line in the same chart showing the yield of oranges and apples respectively")
plt.legend(['Apples','Oranges']);
```

```python
plt.bar(year,yield_oranges)
plt.bar(year,yield_apples,bottom=yield_oranges);
```

# Bar Plot with Averages

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df
```

```python
tips_grouped_df=tips_df.groupby('day')[['tip']].sum()
```

```python
tips_grouped_df
```

```python
plt.bar(tips_grouped_df.day,tips_grouped_df.tip)
```

CAUTION: Note than an index cannot be used in an axis.
    
Hence the index needs to be reset for the above dataframe

```python
tips_grouped_df.reset_index(inplace=True)
```

```python
tips_grouped_df
```

```python
plt.bar(tips_grouped_df.day,tips_grouped_df.tip)
plt.title('Tips given');
```

Advanced Try

```python
tips_grouped_ttlbill_df=tips_df.groupby('day')[['total_bill']].sum()
```

```python
tips_grouped_ttlbill_df.reset_index(inplace=True)
```

```python
tips_grouped_ttlbill_df
```

```python
merged_df = tips_grouped_ttlbill_df.merge(tips_grouped_df,on='day')
```

```python
merged_df
```

```python
plt.bar(merged_df.day,merged_df.total_bill)
plt.bar(merged_df.day,merged_df.tip,bottom=merged_df.total_bill);
```

SNS library has some amazing helper function, that would save time for the users.

using tips_df dataset

```python
sns.barplot('day','total_bill',data=tips_df);
```

The lines cutting each bar represent the amount of variation in the values.

A hue argument to compare bar plots side by side based on a third feature

```python
sns.barplot('day','total_bill',hue='sex',data=tips_df);
```

Bills for male is higher than female

Drawing Horizontal bar charts

This can be done by just switiching the axis

```python
sns.barplot('total_bill','day',hue='sex',data=tips_df);
```

# Heat Map

A heatmap is used to visualize 2 dimensional data like a mar=trix or a table using colors.

The best way to understand it is by looking at an example.

We will use another sample dataset from Seaborn, called "Flights", to visualize monthly passenger footfall at an airport over 12 years.

```python
air_df = sns.load_dataset('flights')
```

```python
air_df
```

We will have to create a pivot before passing this into a heatmap

```python
plt.plot(air_df.passengers)
```

```python
air_df = sns.load_dataset('flights').pivot("month","year","passengers")
```

```python
air_df
```

```python
plt.title("Heatmap of airline traffic wrt to XYZ Airport")
plt.suptitle("No of Passengers in 1000s")
sns.heatmap(air_df);
```

The darker color the lesser footfalls

The footfall at the airport in any given year tends to be higher around Jul and Aug

```python
plt.title("Heatmap of airline traffic wrt to XYZ Airport")

sns.heatmap(air_df,fmt='d',annot=True,cmap='seismic');
```
